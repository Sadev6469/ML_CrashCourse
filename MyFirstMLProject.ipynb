```python
#splitting the data for train dataset and test dataset
#The dataset is a time series data set
#We should always train the model with the past data set and can use the test dataset to predict the future year predictions.
```

```python
train_dataset = teams[teams["year"] < 2012].copy()
test_dataset = teams[teams["year"] >=2012].copy()
```

```python
train_dataset.shape  # train dataset now has 1609 rows and 7 columns
```

```python
test_dataset.shape # testdataset has 405 rows and 7 columns - columns are nothing but features
```

```python
#Training the model
```

```python
from sklearn.linear_model import LinearRegression
# this line creates a linear regression object of LinearRegresssion class and we are naming that object to reg
reg = LinearRegression()
```

```python
predictors = ["athletes","prev_medals"]
target = "medals"
```

```python
#Now we need to tell the reg model to fit using predictors and target
#We are telling the model to use predictors as inputs and medals as label to predict
#This one line does many things
#1. Read the data - the input and the output features to be predicted
#2. Find the best weights and bias 
#3. Minimize the loss using Error Metrics - MAE and MSAE
#4. Find the best fit line and ready to make predictions now.
reg.fit(train_dataset[predictors],train_dataset["medals"])
```

```python
predictions = reg.predict(test_dataset[predictors])# As the model is trained now , We pass the new which is our test data set to make the predictions
```

```python
predictions #Numpy Array of all the predictions
```

```python
test_dataset["predictions"] = predictions
```

```python
test_dataset
```

```python
#Updating the values of the predictions column(values that are less than 0) to 0
#label based indexer in python
test_dataset.loc[test_dataset["predictions"] < 0, "predictions"] = 0
```

```python
#Rounding the values of the prediction column 
test_dataset["predictions"] = test_dataset["predictions"].round()
```

```python
test_dataset # Now we have predictions 
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
error = mean_absolute_error(test_dataset["medals"],test_dataset["predictions"]) # we calculalte MAE b/w these 2 columns
```
